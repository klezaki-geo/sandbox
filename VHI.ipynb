```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use(['dark_background'])
```

```python
# %cd ~
# %cd /home/klemen/_data
# !pwd

# input / output
f_ndvi = '/media/klemen/New Volume/_documents/Projekti/REFARMO/data/2NDVI.csv'
f_lst = '/media/klemen/New Volume/_documents/Projekti/REFARMO/data/2LST.csv'
f_results = '/media/klemen/New Volume/_documents/Projekti/REFARMO/data/2results.csv'
```

```python
# constants for VHI estimation
# https://www.ospo.noaa.gov/Products/land/vvhp/algo.html
a = 0.5
ndvi_min = 0.2
ndvi_max = 0.85
lst_min = -20.
lst_max = 42.
```

```python
# read NDVI
df = pd.read_csv(f_ndvi, encoding='utf-8')
df['date'] = pd.to_datetime(df['system:time_start'], format='%b %d, %Y').dt.strftime("%Y%m%d")
ndvi = df
```

```python
# read LST
df = pd.read_csv(f_lst, encoding='utf-8')
df['date'] = pd.to_datetime(df['system:time_start'], format='%b %d, %Y').dt.strftime("%Y%m%d")
lst = df
```

```python
# merge and interpolate NDVI from 16 to 8 day period
df = pd.merge(lst, ndvi, 'left')
df = df.interpolate(method='linear')
```

```python
# estimate VHI
df['VCI'] = 100. * (df['ndvi_f'] - ndvi_min) / (ndvi_max - ndvi_min)
df['TCI'] = 100. * (lst_max - df['dtcelsius']) / (lst_max - lst_min)
df['VHI'] = a * df['VCI'] + (1 - a) * df['TCI']
```

```python
df = df[['date', 'dtcelsius', 'ntcelsius', 'dif', 'ndvi_f', 'VCI', 'TCI', 'VHI']]
```

```python
df.to_csv(f_results, index=False)
```
